```python
import pandas as pd

# Read data
df = pd.read_csv("events_data.csv")

# Clean data
df = df[df["Club"] != "Unknown Club"]
df = df.loc[df['Description'] != 'No description available']
df.loc[df['Host Organizations'] == 'No organizations found', 'Host Organizations'] = df['Club']
df.loc[df['First Host Org'] == 'No organizations found', 'First Host Org'] = df['Club']
df.loc[df['Categories'] == 'No categories found', 'Categories'] = pd.NA
df['Start Time'] = df['Start Time'].str.replace("PMESTto", "PM EST")
df['Start Time'] = df['Start Time'].str.replace("PMEDTto", "PM EST")
df['Start Time'] = df['Start Time'].str.replace("AMESTto", "AM EST")
df['Start Time'] = df['Start Time'].str.replace("AMEDTto", "AM EST")
df['Start Time'] = df['Start Time'].str.replace("at ", "")
df['Start Time'] = pd.to_datetime(df['Start Time'], format='%A, %B %d %Y %I:%M %p %Z')

df['End Time'] = df['End Time'].str.replace("PMEST", "PM EST")
df['End Time'] = df['End Time'].str.replace("PMEDT", "PM EST")
df['End Time'] = df['End Time'].str.replace("AMEST", "AM EST")
df['End Time'] = df['End Time'].str.replace("AMEDT", "AM EST")
df['End Time'] = df['End Time'].str.replace("at ", "")
df['End Time'] = pd.to_datetime(df['End Time'], format='%A, %B %d %Y %I:%M %p %Z')

# Drop columns and rename
df = df.drop(["Club", "Event Time"], axis=1)

df.rename(columns={
    'Event Name': 'title',
    'Location': 'location',
    'Image URL': 'image',
    'Event URL': 'link',
    'Description': 'details',
    'Host Organizations': 'hosts',
    'First Host Org': 'main_host',
    'Categories': 'categories',
    'Start Time': 'start',
    'End Time': 'end'
}, inplace=True)

# Convert to Unix timestamps (seconds since epoch)
df['start'] = df['start'].astype('int64') // 10**9
df['end'] = df['end'].astype('int64') // 10**9

# Save cleaned data to CSV
df.to_csv("events.csv", index=False)

# Display first 10 rows
df.head(10)
```

```python
from datetime import datetime
import pytz

def seconds_to_minutes_since_midnight(unix_timestamp):
    est = pytz.timezone('US/Eastern')
    dt = datetime.fromtimestamp(unix_timestamp, tz=est)
    return dt.hour * 60 + dt.minute

df['start_day_time'] = df['start'].apply(seconds_to_minutes_since_midnight)
df['end_day_time'] = df['end'].apply(seconds_to_minutes_since_midnight)
```
